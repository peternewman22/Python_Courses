Generating functions: 
You can set default values, but they have to be after the user input argument

```python
def minutes_to_hours(seconds,minutes=70):
    hours = minutes / 60 + seconds / 3600
    return(hours)

print(minutes_to_hours(60))
```

Note, if you need to USE the output of the function, use return.
If you need to just print an output, use print.
You can do both.

```python
def print_sum(num1,num2):
    return(num1+num2)
    print(num1+num2)
print_sum(1,2)
```
