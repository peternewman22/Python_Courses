```python
numbers = [1,2,3]
with open("exercise.txt","w") as file:
    file.seek(0)
    for i in numbers:
        file.write(str(i)+"\n")
```

Note: in the above code, to add any more to the file, you need to change the write mode to a+ and remove the file.seek(0) line
Also, remember, when adding new lines, use "\n" not just \bn
