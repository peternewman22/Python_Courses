### What about a whole column?

```python
df = pandas.read_csv("supermarkets.csv")
df.set_index("ID")
```

First, we need to combine the data in a new column

```python
df["Address"] = df["Address"]+", "+ df["City"]+ ", " + df["State"] + ", " + df["Country"]
df
```

We now want to grab the coordinates

```python
df["Coordinates"] = df["Address"].apply(nom.geocode) #note: you don't have to make a for loop. That's what "apply" does.
```

```python
df["Latitude"] = df["Coordinates"].apply(lambda x: x.latitude if x != None else None) # lambda functions create a temp variable on the fly
df
```

```python
df["Longitude"] = df["Coordinates"].apply(lambda x: x.longitude if x != None else None) # lambda functions create a temp variable on the fly
df
```
