```python
import folium
import pandas
import math
import re

data = pandas.read_excel("GVP_Volcano_List.xlsx",header = 1)

map = folium.Map(tiles="Mapbox Bright")
featureGroup = folium.FeatureGroup(name="Volcanoes")

#Debug
dumpfh = open('out.txt', 'w')

lonData = data["Latitude"]
latData = data["Longitude"]
nameData = data["Volcano Name"]
for lon, lat, name, i in zip(lonData, latData, nameData, range(0, len(nameData))):
    if not math.isnan(lon) and not math.isnan(lat):
        #Debug
        dumpfh.write('{i}: {lon} {lat} {name}\n'.format(i=i, lon=lon, lat=lat, name=name))
        dumpfh.flush()
        
        name = re.sub("'", '', name)
        featureGroup.add_child(folium.Marker(location=[lon,lat],popup=str(name),icon=folium.Icon(color="green")))
    
map.add_child(featureGroup)

map.save("VolcanoMap.html")

#Debug
print('finished')
```

### cleaning up the data

```python
data_raw = pandas.read_excel("GVP_Volcano_List.xlsx",header=1)
data = data_raw.dropna()
# removing names with accents '
#data["Volcano Name"].apply((lambda x: x.replace("'","")),axis=0)
print(list(data["Volcano Name"]))

lon = data["Longitude"]
lat = data["Latitude"]
name = data["Volcano Name"]
elev = data["Elevation"]
```

#### setting up the map

```python
# create the map object
volc_map = folium.Map(tiles="Mapbox Bright")

#add a feature group
all_volcs = folium.FeatureGroup(name="All")



#add markers to the data points
for i,j,k,l in zip(lat,lon,name,elev):
    # take out any ' which would cause a crash
    k = re.sub("'", '', k)
    all_volcs.add_child(folium.Marker(location=[i,j],popup=k+", "+str(l)+"m"))
    # this next line was to islate the problem
    #all_volcs.add_child(folium.Marker(location=[i,j],popup=str(l)))
volc_map.add_child(all_volcs)

#add a control 
folium.LayerControl().add_to(volc_map) 
volc_map

    
```

### Adding layers to the map

```python
sub_volcs_data = data.loc[data["Elevation"]<0]
sub_volcs = folium.FeatureGroup(name="Sub Volcs")


def add_layer(target_map,data,label,color):
    
    lat = data["Latitude"]
    lon = data["Longitude"]
    name = data["Volcano Name"]
    elev = data["Elevation"]
    
    for i,j,k,l in zip(lat,lon,name,elev):
        # take out any ' which would cause a crash
        k = re.sub("'", '', k)
        label.add_child(folium.Marker(location=[i,j],popup=k+", "+str(l)+"m",icon=folium.Icon(color=color)))
    target_map.add_child(label)

add_layer(volc_map,sub_volcs_data,sub_volcs,"green")
folium.LayerControl().add_to(volc_map) 
volc_map
```
