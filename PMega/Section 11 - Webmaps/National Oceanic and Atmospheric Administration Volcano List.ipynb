```python
data_raw = pd.read_html("https://www.ngdc.noaa.gov/nndc/struts/results?type_0=Like&query_0=&op_8=eq&v_8=&type_10=EXACT&query_10=None+Selected&le_2=&ge_3=&le_3=&ge_2=&op_5=eq&v_5=&op_6=eq&v_6=&op_7=eq&v_7=&t=102557&s=5&d=5")
```

```python
data = data_raw[2]
```

```python
data.replace(regex=True,inplace=True,to_replace="'",value="")
```

```python
data.columns
```

```python
volc_map = f.Map(tiles="MapBox Bright")
all_volcs = f.FeatureGroup(name="all")

lat = data["Latitude"]
lon =data["Longitude"]
evel = data["Elev"]
name = data["Volcano Name"]

for la, lo, el, na in zip(lat,lon,elev, name):
    all_volcs.add_child(f.CircleMarker(location=[la,lo],
                                       popup = na+", "+str(el),
                                      radius = 6,
                                      fill=True,
                                      fill_colour = "red",
                                      opacity = 0.7))
```
