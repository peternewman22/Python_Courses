```python
import folium
help(folium.Map)
```

```python
popmap = folium.Map(tiles="Mapbox Bright")
```

```python
import pandas as pd
US_volcs = pd.read_csv("Volcanoes_USA.csv")
```

```python
US_volcs.replace(to_replace="'",value="",regex= True, inplace=True)
```

```python
#create a feature group
US_volc_data = folium.FeatureGroup(name="US Volcanoes")
```

```python
def color_picker(elev):
    if elev > 3000:
        return("red")
    elif 2000 <= elev < 3000:
        return("orange")
    elif 1000 <= elev < 2000:
        return("green")
    elif 0 <= elev < 1000:
        return("dark green")
    elif elev < 0:
        return("blue")
```

```python
# now I populate my map with circler markers
lat = US_volcs["LAT"]
lon = US_volcs["LON"]
elev = US_volcs["ELEV"]
name = US_volcs["NAME"]

# add to the feature group
# note to self... do NOT use an already declared variable as a standin in a for loop
for lt, ln, el, na in zip(lat,lon,elev,name):
    US_volc_data.add_child(folium.CircleMarker(location=[lt,ln], radius=6, popup=na+", "+str(el)+"m", fill_color=color_picker(el),color="grey",fill_opacity=0.7,fill= True))
"""for lt, ln, el in zip(lat,lon,elev):
    US_volc_data.add_child(folium.CircleMarker(location=[lt,ln], radius=6, popup=str(el)+"m", fill_color=color_picker(el),color="grey",fill_opacity=0.7,fill= True))"""
```

```python
popFG = folium.FeatureGroup(name="Population Layer")
popFG.add_child(folium.GeoJson(data=open('world.json', 'r', encoding='utf-8-sig').read(),
style_function=lambda x: {'fillColor':'green' if x['properties']['POP2005'] < 10000000
else 'orange' if 10000000 <= x['properties']['POP2005'] < 20000000 else 'red'}))
```

```python
# now add the featuregroup to the map
popmap.add_child(US_volc_data)
popmap.add_child(popFG)
folium.LayerControl().add_to(popmap)
popmap.save("popmap.html")
```
