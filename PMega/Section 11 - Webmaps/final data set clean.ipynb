```python
data_raw = pd.read_html("http://volcano.oregonstate.edu/volcano_table")
```

```python
data = data_raw[0]
data.replace(to_replace="'",value=" ",regex= True, inplace=True)
```

```python
for i in data['\n          Volcano Name        ']:
    print(i)
```

```python
volc_map = f.Map(tiles="MapBox Bright")
all_volcs = f.FeatureGroup(name="All")  
```

```python
lat = data["\n          Latitude (dd)        "]
lon = data["\n          Longitude (dd)        "]
elev = data["\n          Elevation (m)        "]
name = data["\n          Volcano Name        "]
Type = data["\n          Type        "]    

for la, lo, el, na, ty in zip(lat, lon, elev, name, Type):
    try:
        all_volcs.add_child(f.CircleMarker(location=[la,lo],
                                       radius=6,
                                       color="blue",
                                       fill=True,
                                       fill_color = "red",
                                       popup=na+", "+str(ty)+", "+str(el)+"m",
                                       opacity=0.8))
    except ValueError:
        pass
```

```python
volc_map.add_child(all_volcs)
volc_map.add_child(f.LayerControl())
volc_map
```
