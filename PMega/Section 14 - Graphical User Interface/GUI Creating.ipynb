```python
"""
Three types of widget: Buttons, Entry, Text"""

from tkinter import *

window = Tk() #create a window

def km_to_miles():
    # print(e1_value.get()) # this line is for debugging
    miles = float(e1_value.get())*1.6              #note that .get() pulls out a string
    t1.insert(END,miles
    )
    
b1 = Button(window, # this links my button to the window
           text = "Execute",
           command = km_to_miles # note that you DON'T use brackets
           )
            
            #b1.pack() # this inserts the widget into the window
#alternatively, .grid gives more control over position. Does the same thing as pack.
b1.grid(row=0,
        column=0,
        # rowspan --> allows you to spread a button over multiple rows etc
       ) 


e1_value=StringVar()
e1 = Entry(window, #links to our window
          textvariable=e1_value # stores text entry in e1_value
          ) 
e1.grid(row=0,column=1)



t1 = Text(window, height=1,width=20)
t1.grid(row=0,column=2)


# run the window
window.mainloop() # this goes at the end of the code block
```
