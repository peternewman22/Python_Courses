# Imports

```python
import bokeh 
from bokeh.plotting import figure
from bokeh.io import  output_file, show
```

# Prep some data

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
```

# Prep the Output

```python
# prep output
output_file("Line.html")

#create a figure object
f = figure()
```

## Line Graphs

```python
#create line plot
f.line(x,y)

# show it
show(f)
```

## Triangle

```python
g = figure()
g.triangle(x,y)
show(g)
```

## Circle

```python
h = figure()
h.circle(x,y)
show(h)
```
